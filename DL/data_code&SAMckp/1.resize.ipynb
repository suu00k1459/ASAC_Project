```python
import cv2
import os

def resize_images_in_place(folder_path, target_size):
    # folder_path 내의 모든 파일에 대해 작업 수행
    for filename in os.listdir(folder_path):
        if filename.endswith((".png", ".jpg", ".jpeg")):
            img_path = os.path.join(folder_path, filename)
            img = cv2.imread(img_path)

            # 이미지가 성공적으로 불러와졌는지 확인
            if img is not None:
                # 이미지 크기 조정
                resized_img = cv2.resize(img, target_size)

                # 결과를 동일한 경로에 저장하여 덮어쓰기
                cv2.imwrite(img_path, resized_img)
                print(f"Resized and overwritten image: {img_path}")
            else:
                print(f"Failed to load image: {img_path}")
```

```python
folder_path = '/content/drive/MyDrive/3조/weddingdataset/6 17_off_shoulder/image'
target_size = (560,747)

resize_images_in_place(folder_path, target_size)
```
