```python
import backgroundremover
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("/content/drive/MyDrive/3조/크롤링_강아지데이터셋/테스트 데이터셋/image_main_2_3.jpg")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
# dst2 = np.clip((1+0.5) * image_rgb - 128 * (0.5), 0, 255).astype(np.uint8)

h, s, v = cv2.split(image_rgb)

# 채도를 높이기 위해 값을 조정합니다.
s = cv2.add(s,0)
enhanced_hsv_image = cv2.merge([h, s, v])
enhanced_image = cv2.cvtColor(enhanced_hsv_image, cv2.COLOR_HSV2BGR)
```
