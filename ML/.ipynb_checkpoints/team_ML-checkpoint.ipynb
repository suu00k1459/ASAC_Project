```python
df = ml_df.copy()
```

```python
df.drop(columns = ['Unnamed: 0', 'movie_name', 'genre', 'number of moviegoers'], inplace = True)
```

```python
df.info()
```

```python
y = ml_df.loc[:,'number of moviegoers']
X = df.copy()
```

```python
from sklearn.model_selection import train_test_split

# 훈련 데이터, 평가 데이터, 훈련 데이터 정답, 평가 데이터 정답
X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size = 0.2,
    random_state= 1999
)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
kfold = KFold( n_splits = 5, random_state=1234, shuffle=True)
```

```python
from lightgbm import LGBMClassifier
```

```python
lgbm = LGBMClassifier(n_jobs=-1, random_state=1999)
scores = cross_val_score( lgbm,
                         X, y,
                          cv = kfold, scoring="accuracy")
```
