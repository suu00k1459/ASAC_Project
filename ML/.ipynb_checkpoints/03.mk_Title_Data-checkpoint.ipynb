### 받은 파일로 인덱스 정리

```python
dt1 = pd.read_csv("lst_realfinal.csv")
dt2 = pd.read_csv("box_data_RealFKFKfinal.csv")
```

```python
dt1.drop(['Unnamed: 0','movie_idx'],axis = 1, inplace=True)
dt1['idx'] = [i for i in range(len(dt1))]
```

```python
dt2["movie_nm"] = dt2["image_name"].apply(lambda x : x.replace(".jpg",""))
```

```python
dt2.head()
```

```python
# 변수 생성
dt2["image_size"] = dt2["image_width"]*dt2["image_height"]
dt2["box_size"]= dt2["bbox_width"]*dt2["bbox_height"]
dt2["title_size"] = dt2.groupby("movie_nm")["box_size"].transform('sum')
dt2["title_rate"] = dt2["title_size"]/dt2["image_size"]

for i in range(1,5):
    dt2["label_name"] = dt2["label_name"].apply(lambda x : "Title"+str(i) if (x == "Title "+str(i)) | (x == "title "+str(i))| (x == "title"+str(i)) | (x == "T"+str(i)) else x)
```

```python
def lct(dt): # 박스 위치 확인 변수
    lst= []
    
    for i in range(len(dt)): #데이터 행만큼 
        w = dt.iat[i, 3]/3  # 이미지 넓이 3분할
        h = dt.iat[i, 4]/3  # 이미지 높이 3분할
    
        w_rule = [w,2*w,3*w] 
        h_rule = [h,2*h,3*h]
    
        x_center = (dt.iat[i, 1]+dt.iat[i, 3])/2 # x좌표 값에 width 더한 후 /2
        y_center = (dt.iat[i, 2]+dt.iat[i, 4])/2 # y좌표 값에 heigh 더한 후 /2

        
        lct = 0 # 위치 변수 (3 x 3 행렬이며 좌측 하단부터 0, 우측 상단 8까지 있음) 
        if 0<= x_center < w_rule[0]: # x_center 값이 좌측
            if 0<= y_center < h_rule[0]: # y_center 값이 상단
                lct = 6 
            elif h_rule[0]<= y_center < h_rule[1]: # y_center 값이 중앙
                lct = 3
            else :                       # y_center 값이 하단
                lct = 0
        elif w_rule[0]<= x_center < w_rule[1]: #x_center 값이 중앙
            if 0<= y_center < h_rule[0]:
                lct = 7
            elif h_rule[0]<= y_center < h_rule[1]:
                lct = 4
            else :
                lct = 1
        else:                                  #x_center 값이 우측
            if 0<= y_center < h_rule[0]:
                lct = 8
            elif h_rule[0]<= y_center < h_rule[1]:
                lct = 5
            else :
                lct = 2

        lst.append(lct)
        
    dt["lct"] = lst #새로운 변수로 만듬
    
    return dt # 만들어진 데이터 프레임 반환
    
```

```python
dt = lct(dt2)
dt.head()
```

```python
# 영화별 title size와, title rate
dt3 = dt2[["movie_nm","title_size","title_rate"]]
unique_values_df = dt3.groupby('movie_nm').agg({'title_size': 'unique','title_rate': 'unique'}).reset_index()
unique_values_df["title_size"] = unique_values_df.loc[:,"title_size"].apply(lambda x: x[0])
unique_values_df["title_rate"] = unique_values_df.loc[:,"title_rate"].apply(lambda x: x[0])

# 영화별 
box_size = pd.DataFrame(dt.groupby("movie_nm")["box_size"].sum())
box_size = box_size.reset_index()
box_size = box_size.rename(columns={"box_size":"box_size_sum"})

lct_data = pd.pivot_table(dt, index="movie_nm", columns="label_name",values="lct")
lct_data = lct_data.reset_index().rename_axis(None, axis=1)

box_data = reduce(lambda x, y: pd.merge(x,y, on= "movie_nm", how = "inner"), [unique_values_df,box_size,lct_data])
box_data["idx"] = [i for i in range(len(box_data))]

box_data
```

```python
final_text = pd.merge(dt1, box_data , how = "inner", on ="idx")
final_text["TF"] = final_text["movie_nm_x"] == final_text["movie_nm_y"] 
```

```python
final_text.to_csv("final_text_suyoung_test.csv", header=True, encoding="utf-8")
```
