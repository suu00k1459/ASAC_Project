# 여러 이미지 한번에 분석 후 csv로 저장
for문을 통해 한번에 여러 이미지를 분석

```python
# 파일 불러서 리스트로 저장
import pandas as pd
import numpy as np
import os

directory = r'C:/Users/NT551XCJ/Desktop/MLproject/1693_error'

files = os.listdir(directory)

jpg_files = [file for file in files if file.endswith('.jpg')]

len(jpg_files)
```

```python
jpg_files
```

```python
# jpg_files에 저장된 이미지 이름들을 통해 파일에 있는 이미지를 전부 얼굴 분석함

aws_access_key_id = 'xxxxxxxxxx'
aws_secret_access_key = 'xxxxxxxxxx'
region_name = 'us-east-1'
temp = pd.DataFrame()
error_image = []

for img in jpg_files:
    try:
        # 포스터 분석 jpg_files에 저장된 이미지 이름이 차례로 들어감 (* 경로 맞춰 줄것)
        image_path = f"C:/Users/NT551XCJ/Desktop/MLproject/{img}"


        client = boto3.client('rekognition', aws_access_key_id = aws_access_key_id, aws_secret_access_key =  aws_secret_access_key, region_name = region_name)


        with open(image_path, 'rb') as image_file:
          image_bytes = image_file.read()

        response = client.detect_faces(Image = {'Bytes': image_bytes}, Attributes = ['ALL'])

        # 출력 값 정리
        if len(response['FaceDetails']) != 0:
            # 사람 수 만큼 반복
            for i in range(len(response['FaceDetails'])):
                data = {}
                data['이미지 이름'] = img
                data['배우'] = i
                for key, value in response['FaceDetails'][i].items():
                    if isinstance(value, dict):
                        for sub_key, sub_value in value.items():
                            data[f"{key}_{sub_key}"] = sub_value
                    elif isinstance(value, list):
                        for i, item in enumerate(value):
                            for sub_key, sub_value in item.items():
                                if isinstance(sub_value, str):
                                    key_name = sub_value
                                else:
                                    data[f"{key}_{key_name}_{sub_key}"] = sub_value
                    else:
                        data[key] = value

                # 데이터 행 추가
                temp = pd.concat([temp, pd.DataFrame(data, index=[0])], ignore_index=True)

        else:
            nan_row = pd.DataFrame(np.nan, index=[0], columns=temp.columns)
            nan_row['이미지 이름'] = img
            temp = pd.concat([temp, nan_row], ignore_index=True)
    except:
        print(img)
        error_image.append(img)
        continue

print('Done')

# csv 저장
temp.to_csv("포스터 얼굴 분석.csv",index = False, encoding = 'cp949')
```

# 전처리
배우 별로 한 영화가 여러 행으로 존재했던 데이터 프레임을 영화 한 행으로 통합 변환

```python
import pandas as pd
import numpy as np
```

```python
path = "/content/포스터 얼굴 분석_0529_원본.csv"
data = pd.read_csv(path,encoding='cp949')
data
```

```python
# 배우가 없으면 -1을 채워줌 아래서 10명까지의 정보만 필터링 하기 위함 그냥 nan값인 채 필터링하면 영화 행 자체가 지워짐
data['배우'] = data['배우'].fillna(-1)

# 배우가 최대 10명까지의 정보만 저장
data = data[data['배우'] < 10]
```

```python
# 배우가 없는 행 즉 -1일 경우 NaN으로 다시 변환
data.loc[data['배우'] == -1, '배우'] = np.nan
```

```python
# False는 0으로 True는 1로 변환
data = data.replace('FALSE', 0)
data = data.replace('TRUE', 1)
```

```python
# 성별 여성이면 0 남성이면 1로 변환
data.loc[:,"Gender_Value"] = data.loc[:,"Gender_Value"].apply(lambda x: 0 if x=="Female" else 1)
```

```python
# 0과 1로 잘 변환되었는지 확인
data.loc[:,"Gender_Value"].value_counts()
```

```python
# 빈 값일 경우 nan값으로 채워줌
data = data.fillna(np.nan)
```

```python
# cols_name에 컬럼 명 저장
cols_name = list(data.columns)
cols_name.remove("이미지 이름")
cols_name.remove("배우")
len(cols_name)
```

```python
# 이미지 이름의 유니크 값을 movie_ids에 저장
movie_ids = data.loc[:,"이미지 이름"].unique()
len(movie_ids)
```

```python
# 중복이 제거된 movie_ids를 통해 이미지 이름만 존재하는 데이터 프레임을 만듬
df = pd.DataFrame(
    data= movie_ids,
    columns=["이미지 이름"]
)
df.set_index("이미지 이름", inplace=True)
df.head()
```

```python
# max_people에 배우의 최대 값 저장
max_people = data.loc[:,"배우"].max()
max_people
```

```python
# max_people에 배우의 최대 값을 통해 컬럼들을 미리 만듬 EyeDirection_Confidence_9 까지 만들어짐
# 그리고 값들을 전부 999로 채움

for i in range(0, int(max_people) + 1):
    for j in cols_name:
        new_co_name = f"{j}_{i}"
        df[new_co_name] = 999

print(df.head())
```

```python
df.iloc[:,:20].info()
```

```python
# 999가 들어간 자리에 원래의 값들을 하나씩 넣어줌
for i in range(len(data)):
    m_name = data.iloc[i,0]
    n_people = data.iloc[i,1]
    n_p_data = data.iloc[i,2:]
    if not pd.isna(n_people):
        for j,e_data in enumerate(n_p_data):
            if not pd.isna(e_data):
                c_name = cols_name[j] + "_" + str(int(n_people))
                df.at[m_name, c_name] = e_data
                #print(m_name, df.shape)
            else:
                break
    else:
        continue
print("Done!!")
```
